```python
import numpy as np
import matplotlib.pyplot as plt 
 
# creating the dataset
MSE = [3179.9,2747.63,2454.39,1082.23,1854.72,1694.33,1629.12,1262.74,861.63]
models_label = ['Random Forest Regression','Linear Regression','Logistic Regression','LSTM Lookback=1','LSTM Lookback=5','LSTM Lookback=10'
         ,'LSTM Lookback=20','VAR with Logistic Regression','VAR with LSTM(lookback=1)']
models=[1,2,3,4,5,6,7,8,9]
  
fig = plt.figure(figsize = (25,18))
 
# creating the bar plot
barlist=plt.bar(models, MSE,0.75)
barlist[0].set_color('blue')
barlist[1].set_color('green')
barlist[2].set_color('yellow')
barlist[3].set_color('red')
barlist[4].set_color('magenta')
barlist[5].set_color('cyan')
barlist[6].set_color('orange')
barlist[7].set_color('violet')
barlist[8].set_color('purple')
i=0
for bar in barlist: 
    plt.annotate(format(bar.get_height(), '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=30, xytext=(0, 8), 
                   textcoords='offset points') 
    i=i+1
    
colors = {'Random Forest Regression':'blue', 'Linear Regression':'green','Logistic Regression':'yellow','LSTM Lookback=1':'red','LSTM Lookback=5':'magenta','LSTM Lookback=10':'cyan'
         ,'LSTM Lookback=20':'orange','VAR with Logistic Regression':'violet','VAR with LSTM(lookback=1)':'purple'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]



plt.xlabel("Models used for RUL prediction",fontsize=35)
plt.ylabel("Mean Squared Error",fontsize=35)
plt.title("Comparison of Mean Squared Error of models",fontsize=35)
# plt.legend(handles, labels, loc='best', labelspacing=0.2)
plt.legend(handles, labels,fontsize=20)
plt.grid()

plt.show()
```

```python
MSE = [3179.9,2747.63,2454.39,1082.23,1854.72,1694.33,1629.12,1262.74,861.63]
RMSE = [56.39,52.41,49.54,32.89,43.06,41.16,40.36,35.53,29.35]
MAE = [49.34,44.35,40.5,26.39,33.38,32.39,31.48,28.68,22.91]
error_range = [1,2,3,4,5,6,7,8,9]
models = ['Random Forest Regression','Linear Regression','Logistic Regression','LSTM Lookback=1','LSTM Lookback=5','LSTM Lookback=10'
         ,'LSTM Lookback=20','VAR with Logistic Regression','VAR with LSTM(lookback=1)']
```

```python
fig, ax = plt.subplots(figsize=(18,10))
ax.scatter(error_range,MSE,color='red', label='Mean Squared Error')

fig.suptitle('Comparison of Mean Squared Error of models', fontsize=35)
plt.xlabel("Model used for prediction",fontsize=35)
plt.ylabel("Mean Squared Error",fontsize=35)

ax.legend(loc='upper right')
ax.grid()

for i, txt in enumerate(models):
    ax.annotate(txt, (error_range[i], MSE[i]),fontsize=20)

fig, ax = plt.subplots(figsize=(18,10))
ax.scatter(error_range,RMSE,color='blue', label='Root Mean Squared Error')

fig.suptitle('Comparison of Root Mean Squared Error of models', fontsize=35)
plt.xlabel("Model used for prediction",fontsize=35)
plt.ylabel("Root Mean Squared Error",fontsize=35)

ax.legend(loc='upper right')
ax.grid()

for i, txt in enumerate(models):
    ax.annotate(txt, (error_range[i], RMSE[i]),fontsize=20)

fig, ax = plt.subplots(figsize=(18,10))
ax.scatter(error_range,MAE,color='green', label='Mean Absolute Error')

fig.suptitle('Comparison of Mean Absolute Error of models', fontsize=35)
plt.xlabel("Model used for prediction",fontsize=35)
plt.ylabel("Mean Absolute Error",fontsize=35)

ax.legend(loc='upper right')
ax.grid()

for i, txt in enumerate(models):
    ax.annotate(txt, (error_range[i], MAE[i]),fontsize=20)
```
